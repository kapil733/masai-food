### **extrating int values from Price columns**

```python
def extract_price(price_str):
    price_digits = ''.join(filter(str.isdigit, price_str))
    return int(price_digits) if price_digits else None
```

```python
data1['Price'] = data1['Price'].apply(extract_price)
```
